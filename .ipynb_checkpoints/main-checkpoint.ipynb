```python
import pandas as pd
from IPython.display import Image
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib notebook
```

```python
df = pd.read_csv('./boston.csv')
```

```python
df = df.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "ZN":"PCT_ZONA_RESIDENCIAL",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO_CHARLES",
    "NOX":"OXIDO_NITROSO_PPM",
    "RM":"N_HABITACIONES_MEDIO",
    "AGE":"PCT_CASAS_40S",
    "DIS_EMPLEO":"DISTANCIA_CENTRO_EMPLEO",
    "RAD":"DIS_AUTOPISTAS",
    "TAX":"CARGA_FISCAL",
    "PTRATIO":"RATIO_PROFESORES",
    "B":"PCT_NEGRA",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"
})
df.head()
```

```python
df.dtypes
```

```python
Image('./chart-chooser-data-visualization.png')
```

```python
df.plot.scatter(x='INDICE_CRIMEN', y='VALOR_MEDIANO')
```

```python
sm = scatter_matrix(df,alpha=0.2, diagonal="kde")
```

```python
df.VALOR_MEDIANO.plot.kde()
```

```python
df.VALOR_MEDIANO.plot.hist(xlim=(0,20), bins=20) #histogram
```

```python
#goup by city

value_for_city = df.groupby('CIUDAD')['VALOR_MEDIANO'].mean()
value_for_city.head(10).plot.bar()
```

```python
df.groupby('RATIO_PROFESORES').VALOR_MEDIANO.mean().plot.line()
```

```python
df.RIO_CHARLES.value_counts().plot.pie()
```
