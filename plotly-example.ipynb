```python
import plotly as py
import plotly.graph_objs as go
import numpy as np
```

```python
X = np.random.randint(506, size=1000)
Y = X**2
```

```python
scatter1 = go.Scatter(x=X, y=Y)
scatter2 = go.Scatter(x=X/2, y=Y/2)
py.offline.iplot({
    "data": [scatter1, scatter2],
    "lay"
})
```
